```python
# Bar Charts

oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.bar(years, oscar_revenue, color="red")
plt.title("Revenue in each Year")
plt.xlabel("year")
plt.ylabel("Revenue in $M")

for i in range(len(years)):
    plt.text(years[i], oscar_revenue[i]+10, str(oscar_revenue[i]), ha="center")   # value of revenues, +10 for inc height

plt.ylim(0, max(oscar_revenue)+100)   # inc limit
# plt.xlim(min(years), max(years)+2)
```

```python
import numpy as np

oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]


oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]

years = [2008, 2009, 2010, 2011, 2012]

x = np.arange(len(years))


width = 0.4     # so that no overlapping occurs total width  = 0.8

plt.bar(x-width/2, oscar_revenue, width, label="Oscar Movies")   # shifting to left
plt.bar(x+width/2, non_oscar_revenue, width,label="Non-Oscar Movies")  # shifting to right

plt.title("Oscar vs non_oscar Movies Revenue")
plt.xlabel("Year")
plt.ylabel("Revenue in $M")

plt.legend()           # ❗ to show labels


plt.xticks(x, years)     # Actual year values

plt.show()
```

```python
# Scatter Plots

people = [
    "Person A", "Person B", "Person C", "Person D", "Person E",
    "Person F", "Person G", "Person H", "Person I", "Person J"
]

age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]

bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

colors = ["green" if x < 135 else "red" for x in bp]


plt.scatter(age, bp, s=bp, cmap="OrRd", c=bp)
plt.title("Age vs BP")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")

plt.grid()
plt.colorbar(label="Blood Pressure Values")

for i in range(len(people)):
    plt.annotate(people[i], xy = (age[i], bp[i]), xytext =  (age[i]+1, bp[i]+1))    # To add annotation

plt.xlim(min(age), max(age)+10)
plt.ylim(min(bp), max(bp)+5)
```

```python
# Sub Plots

x = [1,2,3,4,5,6,7,8,9,20]

y1 = [np.sqrt(i) for i in x]
y2 = [i*2 for i in x]
y3 = [i**2 for i in x]
y4 = [i**3 for i in x]

plt.subplot(2, 2, 1)     # 2 x 2 plot = total 4 graphs
plt.plot(x, y1)
plt.title("plot 1 - Square Root")

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title("plot 1 - Double")

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title("plot 1 - Square")

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title("plot 1 - cube")

plt.tight_layout()


```

```python
# Object Oriented API (recommended way)

fig, ax = plt.subplots(2, 2, figsize=(7,7))

ax[0][0].plot(x, y1)
ax[0][0].set_title("plot 1 - Square Root")
ax[0][0].set_xlabel("x-axis value")
ax[0][0].set_ylabel("y-axis value")

ax[0][1].plot(x, y2)
ax[0][1].set_title("plot 2 - Double")

ax[1][0].plot(x, y3)
ax[1][0].set_title("plot 3 - Squares")

ax[1][1].plot(x, y4)
ax[1][1].set_title("plot 4 - Cubes")


fig.suptitle("Multiple Plots")

fig.tight_layout()



```
