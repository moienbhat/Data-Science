```python
import pandas as pd
# load AQI 

df = pd.read_csv("globalAirQuality.csv")

df.describe()
```

```python
# Selecting Data
# Columns
df["city"]    #Print column

df[["city", "aqi"]]    # #Print  multiple columns


# Rows
df.loc[0]     # first   (Label)
df.loc[0:2]    # start index to ending(including it)

df.loc[0]     # first   (index)
df.loc[0:2]     # start label to ending(excluding it)

# Cells (Rows, cols

df.loc[0 ,["city", "aqi"]]
df.columns

df.iloc[0:2, 2]


df.loc[0 ,["city", "aqi","latitude", "longitude"]]

df.iloc[0:3, 1:5]

# cell
df.at[0, "city"]
df.iat[0, 2]

cities = df["city"] # view

cities = df["city"].copy()    # to change data create copy first

cities[0] = "Mumbai"
```

```python
# Filtering Data

df[ df["aqi"] > 100]      # aqi > 100

df[ (df["aqi"] > 100) & (df["temperature"] > 30)]     # two conditions

df[ df["aqi"] > 100] [["city", "aqi"]]      # to print only two cols of city and aqi

aqi_data = df[ (df["aqi"] > 100) & (df["temperature"] > 30)] [["city", "aqi"]]

aqi_data
aqi_data.iloc[0]
aqi_data.loc[6]     # first index has label 6


# Query Method
aqi_val = 100
df.query("aqi > 100 & temperature > 30") [["city", "aqi"]]
df.query("aqi > @aqi_val & temperature > 30") [["city", "aqi"]]       # same as above

aqi_data = df.query("aqi > @aqi_val & temperature > 30") [["city", "aqi"]].copy()   # to change first store copy of that data
```
