```python
# Seaborn
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.set_theme(style="darkgrid")
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    size="size"
)
```

```python
# line Plot

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    kind="line",
    size="size"
)
```

```python
# scatter plot

tips = sns.load_dataset("tips")

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time"
)
```

```python
tips = sns.load_dataset("tips")

sns.lineplot(
    data=tips,
    x="total_bill",
    y="tip"
    
)
```

```python
flights = sns.load_dataset("flights")

sns.lineplot(
    data=flights,
    x="year",
    y="passengers"
)
```

```python
# bar plot

sns.barplot(
    data=tips,
    x="day",
    y="tip",
    hue="sex"
)

```

```python
# Box Plot

sns.boxplot(
    data=tips,
    x="day",
    y="tip"
)
```

```python
# Distribution Plots(Histogram)

penguins = sns.load_dataset("penguins")

sns.histplot(
    data=penguins,
    x="body_mass_g",
    bins=30
)

```

```python
#Matrix Plot(Heat Map)

flights_pivot = flights.pivot(index="month", columns="year", values="passengers")  # to convert flights data to 2D array

sns.heatmap(
    flights_pivot,
    cmap="coolwarm",
    annot=True,
    fmt="d"

)

plt.title("Passengers Heatmap")
plt.tight_layout()
```

```python


fig, ax = plt.subplots()

sns.lineplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex",
    marker="o",
    ax=ax,
    errorbar=None
)

ax.set_title("avg bill value for each day")
ax.set_xlabel("Days")
ax.set_ylabel("Total Bill Amount")

fig.tight_layout()
```
