```python
import pandas as pd
```

```python
info = {
    "Name" : ["Rahul", "Harsh", "Jhon"],
    "CGPA" : [9.5, 7.5, 8.5]
}

df = pd.DataFrame(info)
print(df)
```

```python

s = pd.Series([1,2,3,4,5])
print(s)
print(type(s))

#Index
print(s[2])
```

```python
# Label assignment

s = pd.Series([23,24,25,26], index = ["Adam", "Eve", "Bob", "Charlie"])
print(s)

print(s["Adam"])
print(s[3])
```

```python
#Vectorized Operations

s1 = pd.Series([1,2,3,4,5])
s2 = pd.Series([10,20,30,40,50])

print(s1 + s2)   

# Mutable Values, Immutable size
s1[0] = 100
print(s1)

changed_s1 = s1.drop(0)
print(changed_s1)
```

```python
# DataFrame
info = {
    "Name" : ["Rahul", "Harsh", "Jhon"],     #Dixtionary of lists
    "Age" : [23, 25, 23],
    "CGPA" : [9.5, 7.5, 8.5]
}

df = pd.DataFrame(info)

print(df)

print(df.index)
print(df.columns)
```

```python
df = pd.DataFrame([["Adam", 23], ["Bob", 34], ["Eve", 24]], columns=["Name", "Age"])   # List of Lists
print(df)
```
