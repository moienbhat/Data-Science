```python
# Task

import matplotlib.pyplot as plt

days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

cities = ["New York", "London", "Delhi", "Tokyo"]

temperatures = [
    [22, 23, 21, 24, 25],  # New York
    [18, 19, 17, 20, 21],  # London
    [30, 32, 31, 33, 34],  # Delhi
    [25, 26, 24, 27, 28]   # Tokyo
]

fig, axes = plt.subplots(2,2)


# city_no = 0
# for i in range(2):
#     for j in range(2):
#         axes[i][j].plot(days, temperatures[city_no], marker="o")
#         axes[i][j].set_title(cities[city_no])
#         axes[i][j].grid(True)
#         city_no = city_no + 1

axes = axes.flatten()
for i, ax in enumerate(axes):
    ax.plot(days, temperatures[i], marker='o')
    ax.set_title(cities[i])
    ax.grid(True)


        
fig.suptitle("Temperature in cities over week")
fig.supxlabel("Temperature Values")
fig.supylabel("Days")

plt.grid(True)

fig.tight_layout()

plt.show()
```
